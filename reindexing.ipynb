```python
#import numpy and pandas module
import pandas as pd
import numpy as np

column=['a','b','c','d','e']
index=['A','B','C','D','E']

df1 = pd.DataFrame(np.random.rand(5,5), columns=column, index=index)
df1.head()
```

```python
colum=['a','b','c','g','h']
df1.reindex(colum, axis='columns')
```

```python
df1.reindex(colum, axis='columns', fill_value=1.5)
```

```python
df1.reindex(colum, axis='columns', fill_value='data missing')
```

```python
df1.head()
```

```python
df1.reindex(colum, axis='columns', fill_value='data missing', copy =True)
```
